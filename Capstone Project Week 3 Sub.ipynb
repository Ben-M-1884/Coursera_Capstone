# Explore and Cluster Toronto Neighborhoods

Web scraping Wikipedia to obtain a dataframe containing postcodes, boroughs and neighborhoods in Canada

```python
import pandas as pd
import numpy as np

# Install the Webscraping tool BeautifulSoup4
! pip install beautifulsoup4

# Install lxml the python library for processing XML and HTML
! pip install lxml

# Install requests library for Python. This allows you to send HTTP requests (e.g. GET, POST etc)
# easily.
! pip install requests

# Import Required Modules
from bs4 import BeautifulSoup
import requests

# Obtain and print the formatted HTML code for the wikipedia page containg the list of postal codes in canada
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'lxml')
```

```python
# Create two arrays. One for the Postcode and the other for Borough and Neighborhood Details
# The for loop is used to populate the array
postcode = np.array([])
dist = np.array([])

# Find Table in Webpage and print the details of every canadian postcode in the table
for table in soup.find_all('td'):
    # Find Postcode for item in the table
    Post_code_1 = table.b.text
    #print(Post_code_1)
    # Find Borough and Neighborhood Details
    District = table.find('span').text
    #print(District)  
    
    # Put Postcode into a numpy array
    postcode = np.append(postcode, Post_code_1)
    print(postcode)
    # Put Borough and Neighborhood Details into a numpy array
    dist = np.append(dist, District)
       
```

```python
# Convert two arrays into one dataframe
postcode_dataframe = pd.DataFrame(dist, postcode)
# Reset the Dataframe Index
postcode_dataframe.reset_index(inplace=True)
# Rename the Columns
postcode_dataframe.columns = ['PostalCode', 'Borough/ Neighborhood'] 
# View the dataframe
postcode_dataframe
```

```python
# Create the Series S of the column 'Borough/ Neighborhood' in postcode_dataframe 
# to allow the splitting of strings to seperate the Borough from the neighborhood
s = pd.Series(postcode_dataframe['Borough/ Neighborhood'], dtype="string")

# Remove ')' Character from Neighborhood column
s = s.str.strip(')')

# Split the string on finding the character '(' and split into one new column usinf n=1 and expand = True
s = s.str.split(pat='(', n=1, expand = True)
s.columns = ['Borough', 'Neighborhood'] 
s
```

```python
# Add Two new columns in dataframe 's' to dataframe 'postcode_dataframe'
postcode_dataframe['Borough'] = s['Borough']
postcode_dataframe['Neighborhood'] = s['Neighborhood']

# Delete 'Borough/ Neighborhood' column from postcode_dataframe
del postcode_dataframe['Borough/ Neighborhood']

# Delete all rows that have a 'Not assigned' borough
# Get names of indexes for which column Borough has value 'Not assigned'
indexNames = postcode_dataframe[ postcode_dataframe['Borough'] == 'Not assigned' ].index
# Delete these row indexes from dataFrame
postcode_dataframe.drop(indexNames , inplace=True)

# Combine different Neighborhoods that have the same Postal code
postcode_dataframe = postcode_dataframe.groupby('PostalCode').agg(
    {
        'Borough':'first', 
        'Neighborhood': ', '.join,}
    ).reset_index()
pd.options.display.max_rows = 4000
postcode_dataframe
```

```python
# Explore Neighborhoods in Toronto

# First create a new dataframe that just contains the neighborhoods in totonto 
# from the orginal dataframe  postcode_dataframe
toronto_data = postcode_dataframe[postcode_dataframe['Borough'].str.contains(r'Toronto')].reset_index(drop=True)
toronto_data.head()
```

```python
# Explore all the Neighborhoods in our new dataframe

# Obtain the names of venues, there coordinates within a radius of 500m and limit the number
# of venues to 10
def getNearbyVenues(names, latitudes, longitudes, radius=500, LIMIT=10):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# Perform the above function on each neighborhood in toronto and create a new dataframe
toronto_venues = getNearbyVenues(names=toronto_data['Neighborhood'],
                                   latitudes=toronto_data['Latitude'],
                                   longitudes=toronto_data['Longitude']
                                  )
```

```python
# Dataframe for analysis of each neighborhood by types of venues
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 
# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

# Group Neighborhoods and calculate mean number of each venue type
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()

# Print each neighborhood with the top 5 most common venues in that neighborhood
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# Add print of each neighborhood with the top 5 most common venues in that neighborhood
# to a dataframe

# First sort the venues into descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

# Create New dataframe and list top 10 venues for each neighborhood
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

```python
# K-means clustering will be used to cluster the neighborhoods into 5 clusters
# set number of clusters
kclusters = 5


toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 

```

```python
#
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head() # check the last columns!
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
